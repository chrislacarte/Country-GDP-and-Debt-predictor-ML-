```python
def data_grooming(df):
    df = df.pivot(index='country', columns=df[['subject_descriptor']])
    #rename columns:
    df.columns = ['acc_bal','gross_debt','gov_net_len/borr','gov_rev','gov_expend','gdp_percap_cons','gdp_cons',
                    'gdp_curr', 'gdp_defl', 'gross_savings','inflation_ind','pop','tot_invest', 'unemployment', 
                    'export_g_s', 'import_g_s']
    #arrange order:
    df = df[['acc_bal','gov_rev','gov_expend','gdp_percap_cons','inflation_ind','gov_net_len/borr','tot_invest',
                    'gross_savings','pop', 'unemployment', 
                    'export_g_s', 'import_g_s','gdp_defl','gross_debt','gdp_cons','gdp_curr']]
    #clean commas in 1000s values:
    for i in df.columns:
        df[i] = df[i].str.replace(',', '')
    #transform dtypes to numerics 
    df = df.astype(np.number)
    return df
    
    
def data_cleaning(df):
    df['country'] = df['country'].replace(['Hong Kong SAR',"Korea",'Kosovo','Taiwan Province of China','The Bahamas'], ['Hong Kong','Korea, Republic of','Montenegro','Taiwan, Province of China','Bahamas'])
    df['code'] = df['country'].apply(pc.country_name_to_country_alpha2)
    df['continent'] = df['code'].apply(pc.country_alpha2_to_continent_code)
    df.drop('code', axis=1, inplace=True)
    #rename Kosovo back
    df['country'].replace('Montenegro', 'Kosovo')
    display(df)
    #Spend_class
    #df['spend_class'] = df["gov_expend"].copy()
    labels = ["S1", "S2", "S3","S4"]
    df['spend_class'] = pd.qcut(df["gov_expend"],4, labels = labels)
    display(df)
    #gross_debt_class
    #df['debt_class'] = df["gross_debt"].copy()
    labels = ["0%-5%", "5%-20%", "20%-40%", "40%-60%", "Very High"]
    df['debt_class'] = pd.cut(df['gross_debt'],[0,5,20,40,60,228.397000], labels = labels)
    display(df)
    #invest_class(df):
    #df['invest/gdp_class'] = df["tot_invest"].copy()
    labels = ["0%-20%", "20%-30%", "30%-40%", ">40"]
    df['invest/gdp_class'] = pd.cut(df['tot_invest'],[12,20,30,40,51], labels = labels)
    display(df)
    #reorder columns :
    df = df[['country','continent','spend_class','debt_class','invest/gdp_class','acc_bal','gdp_defl','inflation_ind','gov_rev','gov_expend','gov_net_len/borr','tot_invest','gross_savings', 'export_g_s', 'import_g_s','gross_debt','pop', 'unemployment','gdp_percap_cons','gdp_cons','gdp_curr']]
    display(df)
    
   
    

```
